```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

# Dilation

```python
# Reads in a binary image
image = cv2.imread("./images/j.jpg", 0) 

# Create a 5x5 kernel of ones
kernel = np.ones((5,5), np.uint8)

# Dilate the image
dilation = cv2.dilate(image, kernel, iterations=1)
```

# Erosion

```python
erosion = cv2.erode(image, kernel, iterations=1)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title("dilation")
ax1.imshow(dilation, cmap="gray")

ax2.set_title("original")
ax2.imshow(image, cmap="gray")

ax3.set_title("erosion")
ax3.imshow(erosion, cmap="gray")
```

# Opening

```python
# An opening is erosion followed by dilation. 
# This is useful in noise reduction, in which erosion first gets rid of noise ( and shrinks the object) 
# then dilation enlarges the object again (but the noise will have disappeared from the previous erosion)

opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening, cmap="gray")
```

# Closing

```python
# A Closing is dilation followed by erosion
# This is useful in closing small holes or dark areas wirhin an object

closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing, cmap="gray")
```
